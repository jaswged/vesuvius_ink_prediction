# Loop to Create new dataset

```python
import os
paths = "../data/train/2/surface_volume"
files = os.listdir("../data/train/2/surface_volume")
type(files)
```

```python
height
```

```python
from tqdm import tqdm
for file in tqdm(files, total=len(files)):
    image = cv2.imread(f"../data/train/2/surface_volume/{file}")
    top_image = image[:height // 2, :]
    bottom_image = image[height // 2:, :]
    cv2.imwrite(f"../data/train/4/surface_volume/{file}", top_image)
    cv2.imwrite(f"../data/train/5/surface_volume/{file}", bottom_image)
```

```python
print("Split the labels")
image = cv2.imread("../data/train/2/inklabels.png")
# Split the image into two halves
top_image = image[:height // 2, :]
bottom_image = image[height // 2:, :]
# Save the images
cv2.imwrite(f"../data/train/4/inklabels.png", top_image)
cv2.imwrite(f"../data/train/5/inklabels.png", bottom_image)
```
